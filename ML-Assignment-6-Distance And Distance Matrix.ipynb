```python
#Question:1
#Euclidean Distance.
import math
import numpy as np
import pandas as pd

x1,y1,z1=map(int,input().split())
x2,y2,z2=map(int,input().split())
d=(x1-x2)**2+(y1-y2)**2+(z1-z2)**2
d=math.sqrt(d)
print(d)
```

```python
#Manhattan Distance
x1,y1,z1=map(int,input().split())
x2,y2,z2=map(int,input().split())
d=abs(x1-x2)+abs(y1-y2)+abs(z1-z2)
print(d)
```

```python
#Minkowski Distance
x1,y1,z1=map(int,input().split())
x2,y2,z2=map(int,input().split())
d=(abs(x1-x2))**3 + (abs(y1-y2))**3 +(abs(z1-z2))**3
d=(d)**(1/3)
print(d)
```

```python
#Question:2
def cal(x,y):
    d=0
    for i in range(len(x)):
        d+=(x[i]-y[i])**2
    d=math.sqrt(d)
    return d
def mat(x):
    l=np.zeros((len(x),len(x)))
    for i in range(len(x)):
        for j in range(i,len(x)):
            l[i][j]=cal(list(x[i]),list(x[j]))
            l[j][i]=l[i][j]
    return l
df=pd.read_csv("MLDataSET/iris.csv")
t=list(zip(df['Sepal.Length'],df['Sepal.Width'],df['Petal.Width'],df['Petal.Length']))
print(mat(t))
```

```python
#Question:3
#Scatter Plot
import matplotlib.pyplot as plt 
from sklearn.datasets import load_iris

df = load_iris()
X = df.data
y = df.target
X_ = X[(y == 0) | (y == 1)]
y_ = y[(y == 0) | (y == 1)]
plt.scatter(X_[:, 0], X_[:, 1], c=y_)
plt.xlabel('sepal length (cm)')
plt.ylabel('sepal width (cm)')
plt.show()
```

```python
#Equation Of the Line
from sklearn.linear_model import LogisticRegression

clf = LogisticRegression(random_state=0)
clf.fit(X_[:, :2], y_)
coef = clf.coef_[0]
intercept = clf.intercept_[0]
x_values = np.arange(4, 8, 0.1)
y_values = -(coef[0]*x_values + intercept) / coef[1]
plt.plot(x_values, y_values, '--')
plt.show()

print(f"Equation of the line: Y = {-coef[0]/coef[1]}X + {-intercept/coef[1]}")
```
